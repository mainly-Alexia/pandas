```python
import pandas as pd
import csv as csv
info=['ID','FullName','Gender','M_Status','Nationality','Phone','E_Mail',"Salary"]
with open('employees_info.csv','w') as file:
    myfile=csv.DictWriter(file,delimiter=',',fieldnames=info)
    myfile.writeheader()
```

```python

class add_member:
    def __init__(self):
        while True:
            ID=input("ID_Number:")
            self.ID=ID
            name=input("FullName:")
            self.name=name
            gender=input("gender in full eg Female:")
            self.gender=gender
            M_status=input("M_Status:")
            self.M_status=M_status
            Nationality=input("Nationality:")
            self.Nationality=Nationality
            while True:
                Phone=input("PHONE:")
                if len(Phone)==10 and Phone.isdigit():
                    self.Phone=Phone
                    break
                else:
                    print("the phone number is not exactly 10 numbers!!")
            while True:
                E_mail=input("E-Mail:")
                if '@' in E_mail:
                    self.E_mail=E_mail
                    break
                else:
                    print("your email is not having '@' ")
            salary=input("YOUR salary:")
            self.salary=salary
            option=input("Do you want to add another person 'yess' or 'cancel':").lower()
            data={'ID':self.ID,"FullName":self.name,"Gender":self.gender,"M_Status":self.M_status,"Nationality":self.Nationality,"Phone":self.Phone,"E_Mail":self.E_mail,"Salary":self.salary}
            with open('employees_info.csv','a') as file:
                filewrite=csv.DictWriter(file,delimiter=',',fieldnames=data)
                filewrite.writerow(data)
            if option=="cancel":
                break
```

```python
class RemoveDuplicate:
    def __init__(self):
        df = pd.read_csv('employees_info.csv')
        df = df.drop_duplicates(subset='FullName', keep='first')
        df.to_csv('employees_info.csv', index=False)
        print('All duplicates have been removed, leaving one occurrency')
```
