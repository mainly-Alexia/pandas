```python
import matplotlib.pyplot as plt
```

```python
x=[3,1,3]
y=[3,2,1]
plt.plot(x,y)
plt.legend(['lines'])
s=[2,3,4,5]
r=[7,6,9,0]
plt.plot(s,r)
plt.legend(['fist line','second line'])
plt.show()
```

```python
Fruits=['Apples','Bananas','Cherries',"Dates"]
sales=[400,500,100,50]
plt.bar(Fruits, sales)
plt.title("Fruits sales ")
plt.xlabel("<<<Fruits>>>>")
plt.ylabel("<<<sales>>>")
plt.legend(["star"])
plt.show()
```

```python
label=['Gucci','Versage','Balanciaga','Joker',"Fear of God(FG)","Fendi",]
data=[20,50,12,50,9,70]
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=label)
plt.legend()
plt.show()
```
